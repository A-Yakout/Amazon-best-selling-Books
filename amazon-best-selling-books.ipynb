# Data Analysis ðŸ“‰
**Answering important questions from the Stakeholder**


```python
df_genre=df.groupby('Genre')['Name'].count()
df_genre
```

```python
sns.set_style("whitegrid")
plt.pie(df_genre,labels=['Fiction','Non Fiction'],autopct='%.0f%%');
```

```python
df.groupby('Genre')['User Rating'].mean()
```

Non fiction books are selling better than fiction books ,and the average ratings are better in Non Fiction books

```python
# Relationship of ratings with time.
sns.lineplot(y=df['User Rating'],x=df.Year,hue=df.Genre);
plt.ylabel("Ratings")
plt.xlabel("Years");
```

```python
df_ =df.groupby('Year')['Name'].count()
df_
```

Every year they choose Top 50 books

```python
# Most Selling Books
df_book=df.groupby(by='Name')['est_profit'].sum().sort_values(ascending=False)
df_book.head(5)
```

```python
# Which genre made more sales
genre_sales = df.groupby('Genre')["est_profit"].mean().round(2).reset_index()
genre_sales
```

```python
sns.barplot(data=genre_sales,x=genre_sales['Genre'],y=genre_sales['est_profit'])
```

```python
df_year_rev=df.groupby(by='Year')['est_profit'].sum().sort_values(ascending=False).reset_index()
df_year_rev.head()
```

```python
sns.lineplot(data = df_year_rev,x=df_year_rev.Year,y=df_year_rev.est_profit)
```

2014 is the most selling year 

```python
# Top 5 authors which earned the most
best_authors = df.groupby('Author')['est_profit'].sum().sort_values(ascending=False).head().reset_index()
best_authors
```

```python
sns.barplot(data =best_authors ,x=best_authors.est_profit, y=best_authors.Author )
plt.xlabel('Profit')
plt.title('Top Authors by their profit')
```
